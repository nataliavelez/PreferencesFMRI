# Pilot results

```python
%matplotlib inline

import os
import glob
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from nilearn import image
from nilearn import plotting
from os.path import join as opj
```

Experiment directories:

```python
scratch_dir = os.environ['PI_SCRATCH']
data_dir = opj(scratch_dir, 'OPUS', 'BIDS_data')
deriv_dir = opj(data_dir, 'derivatives')
modelspec_dir = opj(deriv_dir, 'l1model')
model_dir = opj(deriv_dir, 'model')
```

Helper functions:

```python
def load_json(file):
    with open(file, 'r') as handle:
        data = json.load(handle)
        
    return data

def print_con(con):
    n_con = len(con)
    for c in range(n_con):
        print('%i) %s' % (c+1, con[c][0]))
        
def get_con(task, model, subject, con):
    con_dir = opj(model_dir, 'task-%s', 'model-%s', '%s', 'fixedfx', 'con-%02d') % (task, model, subject, con)
    con_file = opj(con_dir, 'zstat1.nii.gz')
    con_img = image.load_img(con_file)
    
    return con_img
```

## Self task

Read contrasts:

```python
self_con_file = opj(modelspec_dir, 'task-self_model-value.json')
self_con = load_json(self_con_file)
print_con(self_con)
```
